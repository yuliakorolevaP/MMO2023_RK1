Для набора данных проведите устранение пропусков для одного (произвольного) числового признака с использованием метода заполнения "хвостом распределения".


**Пропущенные данные в столбце "год" заполнять "хвостом распределения" нецелесообразно.** 

Скорее всего эти пропуски надо оставить или заполнить какой-то заглушкой. Или эти пропуски можно было бы  удалить, так как их доля очень мала в сравнение с выборкой. Поэтому их удаление не повлияет на результаты.

**Тем не менее, в учебных целях заполним пропуски в столбце "год"  "хвостом распределения".**

Вычислим "хвост распределения".

```python
LotFrontage_ev = round(data['Year'].mean() + 1.5*data['Year'].std(),0)
LotFrontage_ev
```

```python
data['Year']=data['Year'].fillna(LotFrontage_ev)
```
