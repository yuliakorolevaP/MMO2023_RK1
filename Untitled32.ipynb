Cреда обучения с подкреплением "Cliff walking" из библиотеки Gym ("Прогулка по скале").

Это стандартная задача с начальным и конечным состояниями и движениями вверх, вниз, вправо и влево. Награда равна -1 за все переходы, кроме тех, которые совершаются в поле с пометкой “Утес”. За вход в это поле агент получает вознаграждение (штраф) в размере -100 и мгновенно возвращается к началу.


**Поле** - матрица 4x12(NumPy):

- [3, 0] - начало слева внизу
- [3, 11] - цель в правом нижнем углу
- [3, 1..10] - утес внизу в центре

Если агент наступает на обрыв, он возвращается в исходное положение. Эпизод завершается, когда агент достигает цели.

**Действия**

Есть 4 дискретных детерминированных действия:

- 0: двигаться вверх
- 1: двигаться вправо
- 2: двигаться вниз
- 3: двигаться влево

**Наблюдения**

Есть 3x12 + 1 возможных состояний. Агент не может быть ни у обрыва, ни у цели (так как это получается в конце эпизода). Остаются все позиции первых 3 строк плюс нижняя левая ячейка. Наблюдение —  текущая позиция.

**Награда**
Каждый шаг - 1 вознаграждение, а шаг в обрыв - -100 вознаграждения.


```python
%matplotlib inline
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
```

```python
''' Зададим поле '''
WORLD_WIDTH = 12
WORLD_HEIGHT = 4

''' Правила '''
START_STATE = (0, 0)
GOAL_STATE = (WORLD_WIDTH-1, 0)

''' Действия '''
ACTION_UP = 0
ACTION_DOWN = 1
ACTION_LEFT = 2
ACTION_RIGHT = 3
ACTIONS = [ACTION_UP, ACTION_DOWN, ACTION_LEFT, ACTION_RIGHT]

''' Политика '''
ALPHA = 0.5
GAMMA = 1
```

```python
#нахождение следующеего состояния и вознаграждение с учетом текущего состояния и действия
next_state_lookup_table = dict()
def get_next_state_and_reward(state, action):
    reward = -1.0
    if state not in next_state_lookup_table:
        next_state_lookup_table[state] = dict()
        
    if action not in next_state_lookup_table[state]:
        x, y = state
        if action == ACTION_UP:
            next_state = (x, min(y+1, WORLD_HEIGHT-1))
        if action == ACTION_DOWN:
            next_state = (x, max(y-1, 0))
        if action == ACTION_LEFT:
            next_state = (max(x-1, 0), y)
        if action == ACTION_RIGHT:
            next_state = (min(x+1, WORLD_WIDTH-1), y)
        
        if (1 <= next_state[0] and next_state[0] <= WORLD_WIDTH-2) and next_state[1] == 0:
            next_state = START_STATE
            reward = -100.0
            
        next_state_lookup_table[state][action] = next_state
        
    return next_state_lookup_table[state][action], reward
```

```python
def one_episode_sarsa(Q, EPSILON):
    timesteps = 0
    trajectory = {}
    sum_of_rewards = 0
    
    # начальное состояние
    state = START_STATE
    
    # выбераем действие, используя  жадную политику
    if np.random.rand() < EPSILON:
        action = np.random.choice(ACTIONS)
    else:
        action = np.argmax(Q[state[0], state[1], :])
        
    while state != GOAL_STATE:

        next_state, reward = get_next_state_and_reward(state, action)
        sum_of_rewards += reward
        
        # выбераем следующее действие, используя  жадную политику
        if np.random.rand() < EPSILON:
            next_action = np.random.choice(ACTIONS)
        else:
            next_action = np.argmax(Q[next_state[0], next_state[1], :]) 
        
        # обновляем Q
        Q[state[0], state[1], action] += ALPHA * \
            (reward + GAMMA * Q[next_state[0], next_state[1], next_action] - Q[state[0], state[1], action])
        timesteps += 1
        
        trajectory[state] = action
        state = next_state
        action = next_action
        
    
    return timesteps, trajectory, sum_of_rewards
```

```python
def one_episode_q_learning(Q, EPSILON):
    timesteps = 0
    trajectory = {}
    sum_of_rewards = 0
    
    # начальное состояние
    state = START_STATE
      
    while state != GOAL_STATE:
       # выбераем действие, используя  жадную политику
        if np.random.rand() < EPSILON:
            action = np.random.choice(ACTIONS)
        else:
            action = np.argmax(Q[state[0], state[1], :])
        
       
        next_state, reward = get_next_state_and_reward(state, action)
        sum_of_rewards += reward
        
        # обновляем Q
        Q[state[0], state[1], action] += ALPHA * \
            (reward + GAMMA * np.max(Q[next_state[0], next_state[1], :]) - Q[state[0], state[1], action])
        timesteps += 1
            
        trajectory[state] = action
        state = next_state
    
    return timesteps, trajectory, sum_of_rewards
```

```python
def print_trajectory(trajectory):
    table = []
    for y in range(WORLD_HEIGHT-1, -1, -1):
        table.append([])
        for x in range(0, WORLD_WIDTH):
            state = (x, y)
            if state == GOAL_STATE:
                table[-1].append('G')
                continue
                
            if (1 <= x and x <= WORLD_WIDTH-2) and y == 0:
                table[-1].append('■■■■')
            elif state in trajectory:
                action = trajectory[state]
                if action == ACTION_UP:
                    table[-1].append('↑')
                elif action == ACTION_DOWN:
                    table[-1].append('↓')
                elif action == ACTION_LEFT:
                    table[-1].append('←')
                elif action == ACTION_RIGHT:
                    table[-1].append('→')
            else:
                table[-1].append('*')
                
    for row in table:
        for ch in row:
            print(ch, end="\t")
        print()
```

```python
def test(method='sarsa', EPSILON=0.5):
    # сбрасываем Q
    Q = np.zeros((WORLD_WIDTH, WORLD_HEIGHT, 4)) # значения пары состояние-действие
    
    n_episode = 500
    ep = 0
    episodes = []
    sums_of_rewards = []
    while ep < n_episode:
        if method == 'sarsa':
            time, trajectory, sum_of_rewards = one_episode_sarsa(Q, EPSILON)
        elif method == 'q-learning':
            time, trajectory, sum_of_rewards = one_episode_q_learning(Q, EPSILON)
        episodes.extend([ep] * time)
        sums_of_rewards.append(sum_of_rewards)
        ep += 1
    
    print('method: %s\n' % method) 
    print('>>>> Траектория на предыдущем эпизоде')
    print_trajectory(trajectory)
    print('\n')
    
    # выводим оптимальную политику
    print('>>>>Оптимальная политика \n') 
    optimal_policy = dict()
    for y in range(WORLD_HEIGHT-1, -1, -1):
        for x in range(0, WORLD_WIDTH):
            state = (x, y)
            if state != GOAL_STATE:
                best_action = np.argmax(Q[x, y, :])
                optimal_policy[state] = best_action
    print_trajectory(optimal_policy)
    print('\n\n')
    
    return episodes, sums_of_rewards
```

```python
def plot_figures(episodes_sarsa, sums_of_rewards_sarsa, 
                 episodes_q_learning, sums_of_rewards_q_learning):
    plt.figure()
    plt.plot(range(len(episodes_sarsa)), episodes_sarsa, label='sarsa', color='red')
    plt.plot(range(len(episodes_q_learning)), episodes_q_learning, label='q-learning', color='blue')
    plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
    plt.title('episodes vs timesteps')
    plt.xlabel('Time steps')
    plt.ylabel('Эпизоды')
    plt.show()


    smoothed_sums_sarsa = []
    for i in range(10, len(sums_of_rewards_sarsa)):
        smoothed_sums_sarsa.append(np.average(sums_of_rewards_sarsa[i:i+10]))

    smoothed_sums_q_learning = []
    for i in range(10, len(sums_of_rewards_q_learning)):
        smoothed_sums_q_learning.append(np.average(sums_of_rewards_q_learning[i:i+10]))

    plt.figure()
    plt.plot(range(len(smoothed_sums_sarsa)), smoothed_sums_sarsa, label='sarsa', color='red')
    plt.plot(range(len(smoothed_sums_q_learning)), smoothed_sums_q_learning, label='q-learning', color='blue')
    plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
    plt.title('Суммарная награда')
    plt.xlabel('Эпизоды')
    plt.ylabel('Саммарная награда')
    plt.show()
```

```python
def test_different_epsilon(EPSILON):
    print('*'*30)
    print('epsilon: %s' % EPSILON)
    print('*'*30)
    print()
    print()
    episodes_sarsa, sums_of_rewards_sarsa = test(method='sarsa', EPSILON=EPSILON)  
    episodes_q_learning, sums_of_rewards_q_learning = test(method='q-learning', EPSILON=EPSILON)
    plot_figures(episodes_sarsa, sums_of_rewards_sarsa, \
                     episodes_q_learning, sums_of_rewards_q_learning)
```

```python
test_different_epsilon(EPSILON=.5)
```

```python
test_different_epsilon(EPSILON=.05)
```

Вывод.
После первоначального переходного процесса Q-learning изучает значения для оптимальной политики, которая проходит прямо по краю обрыва. 

Это приводит к тому, что агент иногда падает со скалы из-за выбора ε  (ε-жадного действия).

Sarsa  принимает во внимание выбор действия и выбирает более длинный, но безопасный путь через верхнюю часть сетки.

Хотя Q-learning действительно изучает значения оптимальной политики, его онлайн-производительность хуже, чем у Sarsa, которая изучает политику обходного пути. Если бы ε постепенно уменьшался, то оба метода асимптотически сходились бы к оптимальной политике.
